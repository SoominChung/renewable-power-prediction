```python
import pandas as pd
import numpy as np

def calculate_overall_metrics(pred, real, capacity, plant_names, min_samples=1):
   """
   전체 data에 대한 메트릭 계산
   """
   # Capacity를 예측/실제값과 같은 shape로 확장
   if len(capacity.shape) == 1:
       capacity_expanded = capacity[:, np.newaxis]
   else:
       capacity_expanded = capacity
   
   # 이용률 계산 (10% 이상만 유효)
   utilization_rate = (real / capacity_expanded) * 100
   valid_mask = utilization_rate >= 10.0
   
   # 오차율 계산 (절대 오차 / 용량 * 100)
   error_rate = np.abs(real - pred) / capacity_expanded * 100
   
   # 유효한 데이터만 선택
   masked_error = np.where(valid_mask, error_rate, np.nan)
   
   # 샘플별 평균 메트릭 (시간축 평균)
   sample_metrics = np.nanmean(masked_error, axis=1)
   
   # 유효한 샘플만 선택 (NaN이 아닌 샘플)
   valid_samples = ~np.isnan(sample_metrics)
   
   if np.sum(valid_samples) == 0:
       return np.nan
   
   # 전체 평균 메트릭
   overall_mean_metric = np.mean(sample_metrics[valid_samples])
   
   return overall_mean_metric

# 각 폴더별 메트릭 계산
results = []

for folder_name in valid_files.keys():
   result_dict = {'folder': folder_name}
   
   # valid error
   try:
       valid_err = calculate_overall_metrics(
           valid_files[folder_name]['pred'], 
           valid_files[folder_name]['real'], 
           valid_files[folder_name]['capacity'], 
           valid_files[folder_name]['plant_names']
       )
       result_dict['valid_err'] = valid_err
   except Exception as e:
       print(f"Error processing {folder_name} valid: {e}")
       result_dict['valid_err'] = np.nan
   
   # test error
   try:
       test_err = calculate_overall_metrics(
           test_files[folder_name]['pred'], 
           test_files[folder_name]['real'], 
           test_files[folder_name]['capacity'], 
           test_files[folder_name]['plant_names']
       )
       result_dict['test_err'] = test_err
   except Exception as e:
       print(f"Error processing {folder_name} test: {e}")
       result_dict['test_err'] = np.nan
   
   # external error
   try:
       external_err = calculate_overall_metrics(
           external_files[folder_name]['pred'], 
           external_files[folder_name]['real'], 
           external_files[folder_name]['capacity'], 
           external_files[folder_name]['plant_names']
       )
       result_dict['external_err'] = external_err
   except Exception as e:
       print(f"Error processing {folder_name} external: {e}")
       result_dict['external_err'] = np.nan
   
   results.append(result_dict)

# DataFrame 생성
metrics_df = pd.DataFrame(results)
sorted_valid_result = metrics_df.sort_values('valid_err')#[['folder','valid_err']]
best_value = sorted_valid_result.iloc[0]['folder']
sec_value = sorted_valid_result.iloc[1]['folder']
print('*****best: ',best_value)
print('*****sec: ',sec_value)

# 전체 결과 출력
print("\n=== All Metrics ===")
sorted_valid_result
```

```python
# =================================================================
# 1. 오류 분석을 위한 샘플 선택 및 플롯
# =================================================================
import os

def get_sample_metrics_with_indices(pred, real, capacity, plant_names):
   """샘플별 메트릭과 인덱스 반환"""
   capacity_expanded = capacity[:, np.newaxis]
   utilization_rate = (real / capacity_expanded) * 100
   valid_mask = utilization_rate >= 10.0
   error_rate = np.abs(real - pred) / capacity_expanded * 100
   masked_error = np.where(valid_mask, error_rate, np.nan)
   sample_metrics = np.nanmean(masked_error, axis=1)
   valid_samples = ~np.isnan(sample_metrics)
   
   return sample_metrics, valid_samples

def plot_sample_analysis(data_type='valid'):
   """특정 데이터 타입에 대한 샘플 분석 플롯"""
   data = best_data[data_type]
   sample_metrics, valid_samples = get_sample_metrics_with_indices(
       data['pred'], data['real'], data['capacity'], data['plant_names']
   )
   
   valid_metrics = sample_metrics[valid_samples]
   valid_indices = np.where(valid_samples)[0]
   
   # 인덱스 선택
   sorted_indices = np.argsort(valid_metrics)
   
   # 최고 오류 5개
   worst_5_idx = valid_indices[sorted_indices[-5:]]
   # 최저 오류 5개  
   best_5_idx = valid_indices[sorted_indices[:5]]
   # 랜덤 5개
   np.random.seed(42)
   random_5_idx = np.random.choice(valid_indices, 5, replace=False)
   
   # 저장 경로 생성
   save_dir = f'/home/local/soominchung_991007/EF/code/tmp/result_analysis/best/{data_type}'
   os.makedirs(save_dir, exist_ok=True)
   
   # 플롯 생성
   categories = [('Best_5_Lowest_Error', best_5_idx), 
                 ('Worst_5_Highest_Error', worst_5_idx),
                 ('Random_5', random_5_idx)]
   
   for cat_name, indices in categories:
       fig, axs = plt.subplots(5, 1, figsize=(12, 15))
       fig.suptitle(f'{data_type.upper()} - {cat_name.replace("_", " ")}', fontsize=16)
       
       for i, idx in enumerate(indices):
           hours = np.arange(24)
           real_vals = data['real'][idx]
           pred_vals = data['pred'][idx]
           capacity = data['capacity'][idx]
           plant_name = location_dict[data['plant_names'][idx]]
           error = sample_metrics[idx]
           
           # 이용률 계산 및 유효성 마스크
           utilization_rate = (real_vals / capacity) * 100
           valid_mask = utilization_rate >= 10.0
           
           # 10% 이상인 점들 (사용되는 점들)
           axs[i].plot(hours[valid_mask], real_vals[valid_mask], 'o-', alpha=0.7, 
                      label='Actual (Used)', color='C0', markersize=4)
           axs[i].plot(hours[valid_mask], pred_vals[valid_mask], 's--', alpha=0.7, 
                      label='Predicted (Used)', color='C1', markersize=4)
           
           # 10% 미만인 점들 (사용되지 않는 점들)
           if np.sum(~valid_mask) > 0:
               axs[i].plot(hours[~valid_mask], real_vals[~valid_mask], 'o', alpha=0.3, 
                          label='Actual (Unused)', color='gray', markersize=3)
               axs[i].plot(hours[~valid_mask], pred_vals[~valid_mask], 's', alpha=0.3, 
                          label='Predicted (Unused)', color='lightgray', markersize=3)
           
           # 10% 이용률 기준선 (빨간 가로선)
           threshold_power = capacity * 0.1
           axs[i].axhline(y=threshold_power, color='red', linestyle='-', alpha=0.7, 
                         linewidth=2, label=f'10% Threshold ({threshold_power:.1f} MW)')
           
           axs[i].set_title(f'{plant_name} ({capacity:.1f} MW) - Error: {error:.2f}%')
           axs[i].set_xlabel('Hour')
           axs[i].set_ylabel('Power (MW)')
           axs[i].set_xticks([0, 6, 12, 18, 23])
           axs[i].legend(fontsize=8)
           axs[i].grid(True, alpha=0.3)
       
       plt.tight_layout()
       
       # 파일 저장
       save_path = os.path.join(save_dir, f'{cat_name}.png')
       plt.savefig(save_path, dpi=300, bbox_inches='tight')
       plt.close()  # 메모리 절약을 위해 닫기
       print(f"Saved: {save_path}")

# 각 데이터 타입별 플롯 실행
for data_type in ['valid', 'test', 'external']:
   plot_sample_analysis(data_type)
```

```python
# =================================================================
# 지역별 성능 차이 통계 분석
# =================================================================

def analyze_regional_performance_statistics(regional_summary_df, plant_results_df):
    """지역별 성능 차이 통계 분석"""
    import scipy.stats as stats
    from scipy.stats import f_oneway, kruskal
    from itertools import combinations
    
    print("=== 지역별 성능 차이 통계 분석 ===")
    
    # 각 data_type별로 분석
    for data_type in ['valid', 'test', 'external']:
        print(f"\n{'='*50}")
        print(f"{data_type.upper()} 데이터 지역별 분석")
        print(f"{'='*50}")
        
        # 해당 data_type의 개별 발전소 데이터 추출
        type_data = plant_results_df[plant_results_df['data_type'] == data_type].copy()
        
        if len(type_data) == 0:
            print(f"No data for {data_type}")
            continue
        
        # 지역별 데이터 분리 (Unknown 제외)
        regions = type_data['광역시'].unique()
        regions = [r for r in regions if r != 'Unknown']
        
        if len(regions) < 2:
            print(f"Not enough regions for statistical analysis (only {len(regions)} regions)")
            continue
        
        print(f"분석 대상 지역: {regions}")
        
        # 지역별 error_rate 데이터 수집
        regional_data = {}
        for region in regions:
            region_errors = type_data[type_data['광역시'] == region]['error_rate'].values
            if len(region_errors) > 0:
                regional_data[region] = region_errors
                print(f"  {region}: n={len(region_errors)}, mean={np.mean(region_errors):.3f}%, std={np.std(region_errors):.3f}%")
        
        # 통계 검정을 위한 데이터 준비
        region_groups = list(regional_data.values())
        region_names = list(regional_data.keys())
        
        if len(region_groups) >= 2:
            # 1. ANOVA 검정
            try:
                f_stat, p_value_anova = f_oneway(*region_groups)
                print(f"\n📊 ANOVA 검정:")
                print(f"  F-statistic: {f_stat:.4f}")
                print(f"  p-value: {p_value_anova:.6f}")
                print(f"  결과: {'지역별 차이 있음' if p_value_anova < 0.05 else '지역별 차이 없음'} (α=0.05)")
            except Exception as e:
                print(f"\n❌ ANOVA 검정 실패: {e}")
            
            # 2. Kruskal-Wallis 검정 (비모수)
            try:
                kw_stat, p_value_kw = kruskal(*region_groups)
                print(f"\n📊 Kruskal-Wallis 검정:")
                print(f"  H-statistic: {kw_stat:.4f}")
                print(f"  p-value: {p_value_kw:.6f}")
                print(f"  결과: {'지역별 차이 있음' if p_value_kw < 0.05 else '지역별 차이 없음'} (α=0.05)")
            except Exception as e:
                print(f"\n❌ Kruskal-Wallis 검정 실패: {e}")
            
            # 3. 사후 검정 (Pairwise comparison) - 유의한 경우만
            if len(region_groups) >= 2 and (p_value_anova < 0.05 if 'p_value_anova' in locals() else True):
                print(f"\n📊 지역간 쌍별 비교 (Mann-Whitney U 검정):")
                
                pairwise_results = []
                for i, j in combinations(range(len(region_names)), 2):
                    region1, region2 = region_names[i], region_names[j]
                    data1, data2 = region_groups[i], region_groups[j]
                    
                    try:
                        statistic, p_value = stats.mannwhitneyu(data1, data2, alternative='two-sided')
                        
                        # Bonferroni 보정
                        n_comparisons = len(list(combinations(range(len(region_names)), 2)))
                        p_adjusted = min(p_value * n_comparisons, 1.0)
                        
                        pairwise_results.append({
                            'region1': region1,
                            'region2': region2,
                            'mean1': np.mean(data1),
                            'mean2': np.mean(data2),
                            'p_value': p_value,
                            'p_adjusted': p_adjusted,
                            'significant': p_adjusted < 0.05
                        })
                        
                        print(f"  {region1} vs {region2}:")
                        print(f"    평균: {np.mean(data1):.3f}% vs {np.mean(data2):.3f}%")
                        print(f"    p-value: {p_value:.6f} (adjusted: {p_adjusted:.6f})")
                        print(f"    결과: {'유의한 차이' if p_adjusted < 0.05 else '차이 없음'}")
                        print()
                        
                    except Exception as e:
                        print(f"  {region1} vs {region2}: 검정 실패 ({e})")
                
                # 가장 유의한 차이들 요약
                significant_pairs = [r for r in pairwise_results if r['significant']]
                if significant_pairs:
                    print(f"🔍 유의한 지역별 차이 ({len(significant_pairs)}개 쌍):")
                    significant_pairs.sort(key=lambda x: x['p_adjusted'])
                    for pair in significant_pairs[:5]:  # 상위 5개만 표시
                        diff = abs(pair['mean1'] - pair['mean2'])
                        print(f"  {pair['region1']} vs {pair['region2']}: {diff:.3f}%p 차이 (p={pair['p_adjusted']:.4f})")
                else:
                    print("🔍 Bonferroni 보정 후 유의한 지역별 차이 없음")
            
            # 4. 효과 크기 (Cohen's f)
            try:
                # 전체 평균
                all_data = np.concatenate(region_groups)
                grand_mean = np.mean(all_data)
                
                # 집단 간 분산
                group_means = [np.mean(group) for group in region_groups]
                group_sizes = [len(group) for group in region_groups]
                
                between_var = sum(n * (mean - grand_mean)**2 for n, mean in zip(group_sizes, group_means)) / sum(group_sizes)
                within_var = np.var(all_data, ddof=1)
                
                cohens_f = np.sqrt(between_var / within_var)
                
                effect_size = "작음" if cohens_f < 0.25 else "중간" if cohens_f < 0.40 else "큰"
                
                print(f"\n📊 효과 크기:")
                print(f"  Cohen's f: {cohens_f:.4f} ({effect_size} 효과)")
                
            except Exception as e:
                print(f"\n❌ 효과 크기 계산 실패: {e}")
    
    # 시각화
    plt.figure(figsize=(20, 12))
    
    for i, data_type in enumerate(['valid', 'test', 'external']):
        type_data = plant_results_df[plant_results_df['data_type'] == data_type].copy()
        
        if len(type_data) == 0:
            continue
        
        # 지역별 박스플롯
        plt.subplot(2, 3, i+1)
        
        regions = [r for r in type_data['광역시'].unique() if r != 'Unknown']
        r_name =[]
        for r in regions:
            r_name.append(province_dict[r])        
        region_data = [type_data[type_data['광역시'] == region]['error_rate'].values for region in regions]
        
        if len(regions) > 1:
            bp = plt.boxplot(region_data, labels=r_name, patch_artist=True)
            
            # 색상 설정
            colors = plt.cm.Set3(np.linspace(0, 1, len(regions)))
            for patch, color in zip(bp['boxes'], colors):
                patch.set_facecolor(color)
                patch.set_alpha(0.7)
        
        plt.title(f'{data_type.upper()} - Regional Error Distribution', fontsize=14, fontweight='bold')
        plt.ylabel('Error Rate (%)', fontsize=12)
        plt.xlabel('Region', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.grid(True, alpha=0.3)
        
        # 바이올린 플롯
        plt.subplot(2, 3, i+4)
        
        if len(regions) > 1:
            parts = plt.violinplot(region_data, positions=range(1, len(regions)+1), showmeans=True)
            
            # 색상 설정
            for pc, color in zip(parts['bodies'], colors):
                pc.set_facecolor(color)
                pc.set_alpha(0.7)
        
        plt.title(f'{data_type.upper()} - Regional Error Density', fontsize=14, fontweight='bold')
        plt.ylabel('Error Rate (%)', fontsize=12)
        plt.xlabel('Region', fontsize=12)

        plt.xticks(range(1, len(regions)+1), r_name, rotation=45, ha='right')
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    # 저장
    save_dir = '/home/local/soominchung_991007/EF/code/tmp/result_analysis/best'
    plt.savefig(os.path.join(save_dir, 'regional_statistical_analysis.png'), dpi=300, bbox_inches='tight')
    plt.close()
    
    print(f"\n📁 결과 저장 완료: {save_dir}/regional_statistical_analysis.png")

# 실행
if 'regional_summary' in globals() and 'plant_results' in globals():
    analyze_regional_performance_statistics(regional_summary, plant_results)
else:
    print("regional_summary 또는 plant_results가 없습니다.")
```

```python
# =================================================================
# 3. 14시 vs 20시 성능 비교
# =================================================================

def compare_14h_vs_20h_performance():
    """14시와 20시 예측 성능 비교"""
    results = []
    
    for data_type in ['valid', 'test', 'external']:
        data = best_data[data_type]
        
        # 데이터 분할
        total_samples = len(data['pred'])
        split_point = total_samples // 2
        
        # 14시 데이터
        pred_14h = data['pred'][:split_point]
        real_14h = data['real'][:split_point]
        capacity_14h = data['capacity'][:split_point]
        plant_names_14h = data['plant_names'][:split_point]
        
        # 20시 데이터
        pred_20h = data['pred'][split_point:]
        real_20h = data['real'][split_point:]
        capacity_20h = data['capacity'][split_point:]
        plant_names_20h = data['plant_names'][split_point:]
        
        # 각각의 메트릭 계산
        metric_14h = calculate_overall_metrics(pred_14h, real_14h, capacity_14h, plant_names_14h)
        metric_20h = calculate_overall_metrics(pred_20h, real_20h, capacity_20h, plant_names_20h)
        
        # 발전소별 성능 비교
        sample_metrics_14h, valid_samples_14h = get_sample_metrics_with_indices(
            pred_14h, real_14h, capacity_14h, plant_names_14h
        )
        sample_metrics_20h, valid_samples_20h = get_sample_metrics_with_indices(
            pred_20h, real_20h, capacity_20h, plant_names_20h
        )
        
        # 유효한 발전소만 비교 (발전소별로 집계)
        valid_both = valid_samples_14h & valid_samples_20h
        
        # 발전소별로 성능 비교
        plant_comparison = {}
        for i in range(len(plant_names_14h)):
            if valid_both[i]:
                plant_name = plant_names_14h[i]
                if plant_name not in plant_comparison:
                    plant_comparison[plant_name] = {
                        'metric_14h': sample_metrics_14h[i],
                        'metric_20h': sample_metrics_20h[i]
                    }
        
        # 발전소별 카운트
        better_14h_plants = 0
        better_20h_plants = 0
        
        for plant_name, metrics in plant_comparison.items():
            if metrics['metric_14h'] < metrics['metric_20h']:
                better_14h_plants += 1
            elif metrics['metric_20h'] < metrics['metric_14h']:
                better_20h_plants += 1
        
        total_plants_compared = len(plant_comparison)
        
        results.append({
            'data_type': data_type,
            'metric_14h': metric_14h,
            'metric_20h': metric_20h,
            'better_14h_plants': better_14h_plants,
            'better_20h_plants': better_20h_plants,
            'total_plants_compared': total_plants_compared,
            'better_14h_ratio': better_14h_plants / total_plants_compared if total_plants_compared > 0 else 0
        })
    
    comparison_df = pd.DataFrame(results)
    print("\n=== 14시 vs 20시 성능 비교 (발전소 기준) ===")
    print(comparison_df.to_string(index=False))
    
    return comparison_df

comparison_results = compare_14h_vs_20h_performance()
```

```python
# =================================================================
# 4. 계절별 성능 분석 (14h + 20h 통합)
# =================================================================

def load_dates_and_analyze_seasons():
    """계절별 성능 분석 (14h, 20h 통합)"""
    import scipy.stats as stats
    from scipy.stats import f_oneway, kruskal
    
    base_path = '/home/local/soominchung_991007/EF/code/tmp/data/normalized_forecast_modeling_data'
    
    # 날짜 데이터 로드
    dates_14h = {}
    dates_20h = {}
    
    for data_type in ['valid', 'test', 'external_test']:
        # 14시 날짜
        date_path_14h = f"{base_path}/14시/{data_type}/dates.npy"
        try:
            dates_14h[data_type] = np.load(date_path_14h, allow_pickle=True)
        except:
            print(f"Could not load {date_path_14h}")
        
        # 20시 날짜
        date_path_20h = f"{base_path}/20시/{data_type}/dates.npy"
        try:
            dates_20h[data_type] = np.load(date_path_20h, allow_pickle=True)
        except:
            print(f"Could not load {date_path_20h}")
    
    def get_season(date_obj):
        """날짜 객체로부터 계절 반환"""
        try:
            # 배열인 경우 첫 번째 원소 사용 (첫 번째 시간)
            if isinstance(date_obj, np.ndarray):
                date_obj = date_obj[0]
            
            # numpy.datetime64인 경우 pandas Timestamp로 변환
            if isinstance(date_obj, np.datetime64):
                date_obj = pd.Timestamp(date_obj)
            # 문자열인 경우 datetime으로 변환
            elif isinstance(date_obj, str):
                date_obj = datetime.strptime(date_obj, '%Y-%m-%d')
            
            month = date_obj.month
            if month in [3, 4, 5]:
                return 'Spring'
            elif month in [6, 7, 8]:
                return 'Summer'
            elif month in [9, 10, 11]:
                return 'Fall'
            else:
                return 'Winter'
        except Exception as e:
            print(f"Error processing date: {type(date_obj)}, error: {e}")
            return 'Unknown'
    
    # 계절별 분석
    seasonal_results = []
    seasonal_data_for_stats = {}  # 통계 분석용 데이터
    
    for data_type in ['valid', 'test', 'external']:
        # 데이터 타입 매핑 (external -> external_test)
        date_key = 'external_test' if data_type == 'external' else data_type
        
        if date_key not in dates_14h or date_key not in dates_20h:
            print(f"Skipping {data_type}: date data not found")
            continue
            
        data = best_data[data_type]
        total_samples = len(data['pred'])
        split_point = total_samples // 2
        
        print(f"\n=== {data_type.upper()} 데이터 처리 중 ===")
        
        # 14h + 20h 데이터 통합
        combined_pred = np.concatenate([data['pred'][:split_point], data['pred'][split_point:]])
        combined_real = np.concatenate([data['real'][:split_point], data['real'][split_point:]])
        combined_capacity = np.concatenate([data['capacity'][:split_point], data['capacity'][split_point:]])
        combined_plant_names = np.concatenate([data['plant_names'][:split_point], data['plant_names'][split_point:]])
        combined_dates = np.concatenate([dates_14h[date_key], dates_20h[date_key]])
        
        print(f"Combined data shape: {combined_pred.shape}")
        print(f"Combined dates shape: {combined_dates.shape}")
        
        # 각 샘플의 계절 분류
        seasons = []
        for date_sample in combined_dates:
            season = get_season(date_sample)
            seasons.append(season)
        
        # 샘플별 메트릭 계산 (통계 분석용)
        sample_metrics, valid_samples = get_sample_metrics_with_indices(
            combined_pred, combined_real, combined_capacity, combined_plant_names
        )
        
        # 유효한 샘플만 선택
        valid_seasons = np.array(seasons)[valid_samples]
        valid_metrics = sample_metrics[valid_samples]
        
        # 통계 분석용 데이터 저장
        seasonal_data_for_stats[data_type] = {
            'seasons': valid_seasons,
            'metrics': valid_metrics
        }
        
        # 계절별 집계
        for season in ['Spring', 'Summer', 'Fall', 'Winter']:
            season_mask = valid_seasons == season
            season_count = np.sum(season_mask)
            
            if season_count == 0:
                continue
            
            season_metrics = valid_metrics[season_mask]
            season_mean = np.mean(season_metrics)
            
            print(f"  {season}: {season_count} samples, mean error: {season_mean:.3f}%")
            
            seasonal_results.append({
                'data_type': data_type,
                'season': season,
                'metric': season_mean,
                'sample_count': season_count,
                'std': np.std(season_metrics)
            })
    
    seasonal_df = pd.DataFrame(seasonal_results)
    print("\n=== 계절별 성능 분석 ===")
    print(seasonal_df.to_string(index=False))
    
    # 통계 분석
    print("\n=== 계절별 성능 통계 분석 ===")
    
    for data_type in ['valid', 'test', 'external']:
        if data_type not in seasonal_data_for_stats:
            continue
            
        seasons = seasonal_data_for_stats[data_type]['seasons']
        metrics = seasonal_data_for_stats[data_type]['metrics']
        
        # 계절별 데이터 분리
        spring_data = metrics[seasons == 'Spring']
        summer_data = metrics[seasons == 'Summer']
        fall_data = metrics[seasons == 'Fall']
        winter_data = metrics[seasons == 'Winter']
        
        season_groups = [spring_data, summer_data, fall_data, winter_data]
        season_names = ['Spring', 'Summer', 'Fall', 'Winter']
        
        # 비어있지 않은 그룹만 선택
        valid_groups = []
        valid_names = []
        for group, name in zip(season_groups, season_names):
            if len(group) > 0:
                valid_groups.append(group)
                valid_names.append(name)
        
        if len(valid_groups) >= 2:
            # ANOVA 검정 (정규성 가정)
            try:
                f_stat, p_value_anova = f_oneway(*valid_groups)
                print(f"\n{data_type.upper()} - ANOVA 검정:")
                print(f"  F-statistic: {f_stat:.4f}")
                print(f"  p-value: {p_value_anova:.4f}")
                print(f"  결과: {'계절별 차이 있음' if p_value_anova < 0.05 else '계절별 차이 없음'} (α=0.05)")
            except:
                print(f"\n{data_type.upper()} - ANOVA 검정 실패")
            
            # Kruskal-Wallis 검정 (비모수)
            try:
                kw_stat, p_value_kw = kruskal(*valid_groups)
                print(f"\n{data_type.upper()} - Kruskal-Wallis 검정:")
                print(f"  H-statistic: {kw_stat:.4f}")
                print(f"  p-value: {p_value_kw:.4f}")
                print(f"  결과: {'계절별 차이 있음' if p_value_kw < 0.05 else '계절별 차이 없음'} (α=0.05)")
            except:
                print(f"\n{data_type.upper()} - Kruskal-Wallis 검정 실패")
            
            # 기술통계
            print(f"\n{data_type.upper()} - 계절별 기술통계:")
            for group, name in zip(valid_groups, valid_names):
                print(f"  {name}: n={len(group)}, mean={np.mean(group):.3f}, std={np.std(group):.3f}")
    
    # 시각화 (계절별 색상)
    season_colors = {
        'Spring': '#90EE90',  # 연두색
        'Summer': '#FFD700',  # 노란색
        'Fall': '#FF8C00',    # 주황색
        'Winter': '#87CEEB'   # 하늘색
    }
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    for i, data_type in enumerate(['valid', 'test', 'external']):
        subset = seasonal_df[seasonal_df['data_type'] == data_type]
        
        if len(subset) > 0:
            # 계절 순서 고정
            seasons_order = ['Spring', 'Summer', 'Fall', 'Winter']
            subset_ordered = []
            colors_ordered = []
            
            for season in seasons_order:
                season_data = subset[subset['season'] == season]
                if len(season_data) > 0:
                    subset_ordered.append(season_data.iloc[0])
                    colors_ordered.append(season_colors[season])
            
            if subset_ordered:
                subset_ordered = pd.DataFrame(subset_ordered)
                
                bars = axes[i].bar(subset_ordered['season'], subset_ordered['metric'], 
                                 color=colors_ordered, alpha=0.8, edgecolor='black', linewidth=1)
                
                # 에러바 추가
                axes[i].errorbar(subset_ordered['season'], subset_ordered['metric'], 
                               yerr=subset_ordered['std'], fmt='none', color='black', capsize=5)
                
                # 샘플 수 표시
                for bar, count in zip(bars, subset_ordered['sample_count']):
                    axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
                                f'n={count}', ha='center', va='bottom', fontsize=10, fontweight='bold')
                
                axes[i].set_title(f'{data_type.upper()} - Seasonal Performance', fontsize=14, fontweight='bold')
                axes[i].set_ylabel('Error Rate (%)', fontsize=12)
                axes[i].set_xlabel('Season', fontsize=12)
                axes[i].grid(True, alpha=0.3)
                axes[i].tick_params(axis='both', labelsize=11)
        else:
            axes[i].set_title(f'{data_type.upper()} - No Data', fontsize=14)
    
    plt.tight_layout()
    
    # 저장
    save_dir = '/home/local/soominchung_991007/EF/code/tmp/result_analysis/best'
    plt.savefig(os.path.join(save_dir, 'seasonal_analysis_combined.png'), dpi=300, bbox_inches='tight')
    plt.close()
    
    # CSV 저장
    seasonal_df.to_csv(os.path.join(save_dir, 'seasonal_results_combined.csv'), index=False)
    
    print(f"\n결과 저장 완료: {save_dir}")
    
    return seasonal_df

# 계절별 분석 실행
seasonal_results = load_dates_and_analyze_seasons()
```
