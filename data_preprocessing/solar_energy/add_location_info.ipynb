```python
data_path = '../../data/processed_data'
meta_data = pd.read_csv(os.path.join(data_path, 'meta_data.csv'))

# 먼저 발전소 이름과 location_name을 매핑한 dict 생성
mapping = {}
for _, row in df.iterrows():
    for plant in row["발전소"]:
        mapping[plant.strip()] = row["location_name"]

# meta_data의 name과 mapping 키를 비교해서 location_name 할당
meta_data["location_name"] = meta_data["name"].map(mapping)

meta_data = pd.merge(meta_data,df[['location_name','광역시']],on='location_name',how='left')#.drop_duplicates(subset=['location_name']).to_csv(os.path.join(data_path, 'location_mapping.csv'), index=False)

meta_data
```
